<h1>PART II</h1>
<h2>Build Decision Tree(ID3)</h2>


```python
#construct decision tree in recursive fashion
def build(data,labels):
    
    '''
    training_label = []
    for value in data:
        training_label.append(value[-1])
    '''

    training_label = [value[-1] for value in data]
    #when terminating recurrence
    if training_label.count(training_label[0]) == len(training_label):
        return training_label[0]
    
    best_feature = which_to_split(data)
    feature_label = labels[best_feature]
    tree = {feature_label:{}}
    labels.remove(feature_label)
    '''
    features = []
    for record in data:
        features.append(record[best_feature])
    feature_val = set(features)
    '''
    feature_val = set([record[best_feature] for record in data])
       
    for value in feature_val:
        sub_labels = labels[:]
        #build sub-trees for each child
        tree[feature_label][value] = build(split_data(data, best_feature, value),sub_labels)
    
    return tree
```

```python
#calculate entropy
import math
def cal_entropy(data):
    entropy = 0
    
    #construct a dictionary to store counting results of labels 
    entropy_labels = {}
    
    for record in data:
        label = record[-1]
        if label not in entropy_labels.keys():
            entropy_labels[label] = 1
        else:
            entropy_labels[label] += 1
   
    for key in entropy_labels.keys(): 
        prob = float(entropy_labels[key]) / len(data)
        entropy = -prob*math.log(prob,2)
    
    return entropy
'''
initial version:
    entropy = 0
    entropy_labels = {'yes':0,'No':0}
    for record in data:
        label = record[-1]
        entropy_labels[label] += 1
    for key in entropy_labels.keys(): 
        prob = float(entropy_labels[key]) / len(data)
        entropy = -prob*math.log(prob,2)
    return entropy
'''
```

```python
#split data when feature[n] == value
def split_data(data,n,value):
    result = []
    
    for record in data:
        if record[n] == value:
            trimmed_record = record[:n]
            trimmed_record.extend(record[n+1:])
            result.append(trimmed_record)
    
    return result
```

```python
#chose the best feature(highest info gain) to split
#return best_feature_index
def which_to_split(data):
    #get base entropy
    entropy = cal_entropy(data)
    max_info_gain = 0
    best_feature_index = -1
    
    for i in range(len(data[0])-1):
        #extract distinct feature list to iterate through
        '''
        features = []
        for record in data:
            features.append(record[i])
        features = set(features)
        '''
        features = set([record[i] for record in data])
        new_entropy = 0
        
        #calculate new entropy
        for value in features:
            sub_data = split_data(data, i, value)
            new_entropy += (len(sub_data)/float(len(data))) * cal_entropy(sub_data)
        
        #calculate infomation gain
        info_gain = entropy - new_entropy
        
        if info_gain > max_info_gain:
            max_info_gain = info_gain
            best_feature_index = i
   
    return best_feature_index
```

```python
def predict(tree, test_data, labels):
    splitter = tree[tree.keys()[0]]
    label_index = labels.index(tree.keys()[0])
    
    for key in splitter.keys():
        if test_data[label_index] == key:
            #go through the tree untill no more sub trees
            if str(type(splitter[key])) == "<type 'dict'>":
                predicted_class = predict(splitter[key],test_data,labels)
            else: predicted_class = splitter[key]
    
    return predicted_class
```

<h1>PART III</h1>
<h2>Run Decision Tree(ID3)</h2>

```python
#Parse raw data file: "dt-data.txt"

from string import split
labeled_data=[]

for line in open('dt-data.txt'):
    trimmed_line=''
   
    for char in line:
        if char.isalpha() or char == ',':
            trimmed_line+=char
    
    trimmed_line_split = split(trimmed_line,',')
    
    #delete empty line
    if len(trimmed_line_split) > 1:
        '''
        for value in trimmed_line_split:
            labeled_data.append([value])
        '''
        labeled_data.append([value for value in trimmed_line_split])

labels = labeled_data[0]
data = labeled_data[1:]
```

```python
tree = build(data,labels)
```

```python
test_data = ['Large','Moderate','Cheap','Loud','CityCenter','No','No']
labels = labeled_data[0]
```
