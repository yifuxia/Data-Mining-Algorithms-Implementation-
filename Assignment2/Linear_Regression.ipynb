```python
#independent data X
X  = np.delete(data, np.s_[2:3], axis=1)
#add constant
X = np.append(X,(np.zeros(len(X)) + 1).T.reshape(50,1),axis=1)
X
```

```python
#dependent data Y
Y = np.delete(data, np.s_[0:2], axis=1)
Y
```

```python
beta = np.zeros(X.shape[1])
beta
```

```python
def h(beta,X):
    return X.dot(beta.T)
```

```python
def calc_cost(beta,X,Y):
    return np.mean(Y - h(beta,X))**2
```

```python
def gd(beta, X, Y, rate=.0001, converge=.0001):
    cost_iter = []
    cost = calc_cost(beta, X, Y)
    cost_iter.append([0, cost])
    cost_diff = 1
    i = 1
    while(cost_diff > converge):
        old_cost = cost
        beta = beta - (rate * X.T .dot ( X.dot(beta) - Y.squeeze()))
        cost = calc_cost(beta, X, Y)
        cost_iter.append([i, cost])
        cost_diff = old_cost - cost
        i+=1
    return beta, ("iterations: ",i)
```

```python
gd(beta, X, Y,0.0001,0.1)
```
