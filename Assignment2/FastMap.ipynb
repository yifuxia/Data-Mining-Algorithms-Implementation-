```python
import csv
import math
import numpy as np
import random
from matplotlib import pyplot as plt
from IPython.display import Image
data = []
with open('dims.txt') as csvfile:
    points = csv.reader(csvfile,delimiter=',')
    for point_str in points:
        data.append(point_str)
data=np.array(data).astype(np.float)

data
```

```python
def distance(X,Y):
    d = 0
    if len(X) != len(Y):
        print 'error!'
    else:
        for index,x in enumerate(X):
            d += (X[index] - Y[index])**2
    return math.sqrt(d)            
```

```python
def choose_distant_objects(O,D,iterations = 5):
    b = random.randint(0,N-1)
    while iterations > 0:
        a = farthest(b,D)
        b = farthest(a,D)
        iterations -= 1
    return (a,b)
```

```python
def farthest(index,D):
    distance = -1
    farthest_index = -1
    for i in range(len(D[index])):
        if D[index][i] > distance:
            distance = D[index][i]
            farthest_index = i
    return farthest_index
```

```python
def FastMap(k,D,O):
    global X
    global col_num
    global PA
    (a,b) = choose_distant_objects(O,D)
    PA[0][col_num] = a
    PA[1][col_num] = b
    print PA
    if D[a][b] == 0:
        for row in X:
            row[col_num] = 0
        return
    else:
        for index,obj in enumerate(X):
            obj[col_num] = (D[a][index]**2 +
                           D[a][b]**2   -
                           D[b][index]**2) / (2* D[a][b])
    if k-1 <= 0:
        return
    else:
        new_D = np.zeros((N,N))
        for i in range(N):
            for j in range(N):
                xi = (D[a][i]**2 +
                               D[a][b]**2   -
                               D[b][i]**2) / (2* D[a][b])
                xj = (D[a][j]**2 +
                               D[a][b]**2   -
                               D[b][j]**2) / (2* D[a][b])
                new_D[i][j] = math.sqrt(D[i][j]**2 -np.round(xi-xj,5)**2)
        col_num += 1
        FastMap((k-1),new_D,O)
```
