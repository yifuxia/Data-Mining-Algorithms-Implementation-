```python
import csv
import math
import numpy as np
import random
from matplotlib import pyplot as plt
from IPython.display import Image
data = []
with open('dims.txt') as csvfile:
    points = csv.reader(csvfile,delimiter=',')
    for point_str in points:
        data.append(point_str)
data=np.array(data).astype(np.float)
data
```

```python
#Compute the mean vector
mean_v =[]
for i in range(len(data.T)):
    mean_v.append(np.mean(data.T[i]))
```

```python
mean_v
```

```python
#Compute the covariance matrix
cov_mat = np.cov(data.T)
```

```python
cov_mat
```

```python
#Compute eigenvalue and eigenvector of cov_mat
eig_val, eig_vec = np.linalg.eig(cov_mat)
```

```python
eig_val, eig_vec
```

```python
#Choose 2 eigenvectors with the largest eigenvalues
eig_pairs = [(np.abs(eig_val[i]), eig_vec[:,i]) for i in range(len(eig_val))]
```

```python
eig_pairs.sort()
eig_pairs.reverse()
```

```python
W =[]
W.append(eig_pairs[0][1].T)
W.append(eig_pairs[1][1].T)
```

```python
W = np.array(W)
```

```python
W
```

```python
dim_reduced_data = W.dot(data.T).T
np.set_printoptions(suppress=True)
dim_reduced_data
```
